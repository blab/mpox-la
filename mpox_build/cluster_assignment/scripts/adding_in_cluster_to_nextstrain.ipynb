```python
import pandas as pd
```

```python
cluster_meta = pd.read_csv("../../../multitree_coalescent/data/la_clusters_with_metadata_11_07.tsv", sep = "\t")
original_meta = pd.read_csv("../../phylogenetic/data/metadata.tsv", sep = "\t")
```

```python
original_meta
```

```python
combined_df = pd.merge(cluster_meta, original_meta, left_on = "strain", right_on = "accession", how = "left")
combined_df = combined_df[["strain_x", "strain_y", "cluster"]]
```

```python
combined_df = combined_df.rename(columns = {"strain_x": "accession", "strain_y": "strain"})
combined_df = combined_df.set_index("strain")
```

```python
combined_df.to_csv("cluster_metadata_for_nextstrain.tsv", sep = "\t")
```
