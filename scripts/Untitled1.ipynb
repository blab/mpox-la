```python
import pandas as pd
import numpy as np
```

```python
filename = "../../oropouche/phylogenetic/test/metadata_diff.tsv"
df = pd.read_csv(filename, sep = "\t")
```

```python
df
```

```python
# Get the total number of rows
num_rows = df.shape[0]

# Loop through the DataFrame in chunks of 10 rows
for i in range(0, num_rows, 15):
    # Extract the chunk of 10 rows (or fewer for the last chunk)
    chunk = df.iloc[i:i+15]
    
    # Define the file name for each chunk
    file_name = f'../../oropouche/phylogenetic/test/excluded_seqs_{i//15 + 1}.tsv'
    
    # Save the chunk as a TSV file
    chunk.to_csv(file_name, sep='\t', index=False)
    
    print(f"Saved: {file_name}")
```
